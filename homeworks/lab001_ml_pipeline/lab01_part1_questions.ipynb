## Question 3

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary (hyperplane between classes in feature space) of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

*Your Answer:*
4

*Your Explanation:*

1. False. The decision boundary of k-NN is non-linear. It is determined locally by the distance to training points, resulting in a complex, jagged shape rather than a single straight line (hyperplane). For instance $k=1$, the boundary is composed of piecewise linear segments (forming Voronoi tessellations) that surround the training examples, which is not a linear hyperplane.

2. False. For $k=1$ the training error always = 0. So we need to find some way to reach this trainig error with $k=5$. For instance if the data is well-separated. In that case, the errors are equal ($0=0$). Therefore, the 1-NN error is less than or equal ($\le$) to 5-NN, but not always lower ($<$).

3. False. 1-NN is prone to noise and outliers from the training data. This often leads to higher test error compared to 5-NN, which smooths out the decision boundaries. Since 1-NN does not always have lower test error, the statement is false.

4. True. The computational complexity of classifying a test example with a naive k-NN implementation is $O(N \cdot \log k) = O(N)$, where $N$ is the size of the training set. The algorithm needs to compare the test sample against all $N$ training samples, so the time increases as the dataset grows.

5. False. Because of 4 is True
